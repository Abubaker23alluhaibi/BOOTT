```
import numpy as np
import matplotlib.pyplot as plt
#210513414
def lin(x): return x
def thr(x): return np.where(x>=0.5,1.0,0.0)
def sig(x): return 1.0/(1.0+np.exp(-x))

def init_rand(s, seed=123):
    g=np.random.default_rng(seed)
    return [g.uniform(0.0,2.0,(s[i],s[i+1])) for i in range(len(s)-1)]
def fpass(x, W, act):
    a=x.reshape(1,-1)
    for Wi in W:
        a=act(a@Wi)
    return a.ravel()
def sweep_plot(name, s, W, fix, idx, xs, act, act_name):
    Y=[]
    for v in xs:
        x=fix.copy(); x[idx]=v
        Y.append(fpass(x,W,act))
    Y=np.vstack(Y)
    plt.figure(figsize=(7,4))
    for j in range(Y.shape[1]):
        plt.plot(xs, Y[:,j], label=f"y{j+1}")
    plt.title(f"{name} - {act_name}")
    plt.xlabel(f"x{idx+1}")
    plt.ylabel("out")
    plt.grid(True); plt.legend(); plt.show()

s1=[3,5,5,5,5,3]
W1=init_rand(s1, seed=123)
x1=np.array([0.5,1.2,0.8])

for n,act in [("Linear",lin),("Threshold",thr),("Sigmoid",sig)]:
    print("Model-1", n, fpass(x1,W1,act))

xs=np.linspace(-10,10,201)
fix1=np.array([0.0,0.5,1.0])

for n,act in [("Linear",lin),("Threshold",thr),("Sigmoid",sig)]:
    sweep_plot("Model-1 (3-5-5-5-5-3)", s1, W1, fix1, 0, xs, act, n)
```
